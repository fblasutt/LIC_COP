# Dynamic Collective Limited Commitment Model
In this simple example, singlehood is absorbing and couples choose
individual consumption, $c_{j,t}$ for $j\in\{w,m\}$ and home goods consumption $Q_t$,
$c_{t}$. Individual preferences are as follows,
\begin{align*}
U_{j}(c_{j,t},Q_{t}) & =\frac{1}{1-\rho_{j}}\left(\alpha_{1,j}c_{j,t}^{\phi_{1}}+\alpha_{2,j}Q_{t}^{\phi_{1}}\right)^{\phi_{2}}.
\end{align*}
Home goods $Q_t$ are produced using time and money:
$$
Q_t=(\theta x_t^\lambda+(1-\theta)*(2\overline{t}+(1-\overline{t})*P^w_t)^\lambda)^{1/\lambda},
$$
where $x_t$ are expenditures, $\overline{t}$ is a (fixed) time devoted to producing the home goods if j participates in the market, and $P_{w,t}$ indicates women's labor force participation. The household budget for a couple is
$$
A_{t}+x_{t}+c_{w,t}+c_{m,t}=RA_{t-1}+y_{w,t}P_{w,t}+y_{m,t}
$$
where $A_{t-1}$ is the beginning of period $t$ wealth. Labor income $y_{j,t}$ follows:
$$
\log(y_{j,t})=\text{Age polynomial}_{j,t}+z_{j,t}+h_{j,t}, \text{where } z_{j,t}=z_{j,t-1}+\epsilon_{j,t},
$$
where $\epsilon_{j,t}\sim iid \mathcal{N}(0,\sigma_{j}^{2})$. Human capital $h_{j,t}=0$ if $j=m$, while for women human capital $h_{w,t+1}\in\[H_0,H_1\]=\{0,-\tau\}$ evolves according to the following transition matrixes, which make future human capital $h_{w,t+1}$ depending on past human capital $h_{w,t}$ and labor supply decisions in $t$, $P^w_t$=



\begin{equation}
  \begin{cases}
    \begin{bmatrix}
1 & 0 \\
1-Pr & Pr 
\end{bmatrix}, & \text{if $P^w_t=1$}.\\
    \begin{bmatrix}
1-Pr & Pr \\
0   & 1 
\end{bmatrix}, & \text{if $P^w_t=0$}.
  \end{cases}
\end{equation}



In each period, the couple receives
a random value of remaining as a couple, $\psi_{t}$, which follows
a unit-root process,
$$
\psi_{t+1}=\psi_{t}+\varepsilon_{t+1}
$$
where $\varepsilon_{t}\sim iid \mathcal{N}(0,\sigma_{\psi}^{2})$.
This love-shock is the only source of uncertainty in the model. The
state variables in $t$ for a couple is then $\mathcal{S}_{t}=(\psi_{t},A_{t-1},z_{w,t},z_{m,t},h_{w,t},\mu_{t-1})$.

**The value of divorcing for a couple** is then
$$
V_{j,t}^{m}(\mathcal{S}_{t})=D_{t}^{\star}V_{w,t}^{m\rightarrow s}(z_{w,t},h_{w,t}, \kappa_{w} A_{t-1})+(1-D_{t}^{\star})V_{m,t}^{m\rightarrow m}(z_{m,t},h_{m,t}, \kappa_{m} A_{t-1})
$$
where $\kappa_{j}$ is the share of household wealth member $j$ gets
in case of divorce. The choice to divorce,
$D_{t}^{\star}$, is discussed below. Note that $h_{m,t}=0 \ \forall t$

**The value of transitioning into singlehood** is
\begin{align*}
V_{j,t}^{m\rightarrow s}(z_{j,t},h_{j,t}, A_{t-1}) & =\max_{c_{j,t},x_{t}}U_{j}(c_{j,t},Q_{t})+\beta V_{j,t+1}^{s}(z_{j,t+1},h_{j,t+1}, A_{t})\\
 & \text{s.t.}\\
A_{t} & =RA_{t-1}+y_{j,t}-x_{t}-c_{j,t}
\end{align*}
where $V_{j,t+1}^{s}(z_{j,t+1},h_{j,t},A_{t})=V_{j,t+1}^{m\rightarrow s}(z_{j,t+1},h_{j,t},A_{t})$ since
singlehood is absorbing.

**The value of remaining married** is
\begin{align*}
V_{j,t}^{m\rightarrow m}(\mathcal{S}_{t}) & =U_{j}(c_{j,t}^{\star},Q_{t})+\psi_{t}+\beta\mathbb{E}_{t}[V_{j,t+1}^{m}(\mathcal{S}_{t+1})]\\
 & \text{s.t.}\\
A_{t} & =RA_{t-1}+P_{w,t}^{\star}y_{w,t}+y_{m,t}-(x_{t}^{\star}+c_{w,t}^{\star}+c_{m,t}^{\star})
\end{align*}
where $(x_{w,t}^{\star},c_{m,t}^{\star},c_{t}^{\star},P_{w,t}^{\star})$ and $\mu_{t}$
are found along with $D_{t}^{\star}$ in the following way.

Let the solution to a problem of couples, under the condition that
they remain together taking the bargaining power, $\mu$, be
\begin{align*}
\tilde{c}_{w,t}(\mu),\tilde{c}_{m,t}(\mu),\tilde{x}_{t}(\mu), \tilde{P}_{w,t}(\mu)  & =\arg\max_{c_{w,t},c_{m,t},x_{t},P_{w,t}}\mu v_{w,t}(\psi_{t},z_{w,t},z_{m,t},h_{w,t},A_{t-1},c_{w,t},c_{m,t},x_{t},P_{w,t},\mu)\\
 & +(1-\mu)v_{m,t}(\psi_{t},z_{w,t+1},z_{m,t+1},h_{w,t},A_{t-1},c_{w,t},c_{m,t},x_{t},P_{w,t},\mu)\\
 & \text{s.t.}\\
A_{t} & =RA_{t-1}+P_{w,t}y_{w,t}+y_{m,t}-(x_{t}+c_{w,t}+c_{m,t})
\end{align*}
where the value-of-choice given some $\mu$ is
\begin{align}
v_{j,t}(\psi_{t},z_{w,t},z_{m,t},h_{w,t},A_{t-1},\mu,c_{w,t},c_{m,t},x_{t},P_{w,t}) & =U_{j}(c_{j,t},Q_{t})+\psi_{t}+\beta\mathbb{E}_{t}[V_{j,t+1}^{m}(\mathcal{S}_{t+1})].
\end{align}

First, solve the unconstrained problem under the assumption that none
of the participation constraints are violated, such that $\mu=\mu_{t}=\mu_{t-1}$.
This gives $\tilde{c}_{w,t}(\mu_{t-1}),\tilde{c}_{m,t}(\mu_{t-1}),\tilde{x}_{t}(\mu_{t-1}),\tilde{P}_{w,t}(\mu_{t-1})$
and individual values of marriage as $v_{j,t}(\psi_{t},z_{w,t},z_{m,t},h_{w,t},A_{t-1},\mu_{t-1},\tilde{c}_{w,t}(\mu_{t-1}),\tilde{c}_{m,t}(\mu_{t-1}),\tilde{x}_{t}(\mu_{t-1}))$. 

**Checking the Participation Constraint**
Secondly, check the participation constraints for three cases. For
this purpose let 
$$
S_{j,t}(\mu)=S_{j,t}(\psi_{t},z_{w,t},z_{m,t},h_{w,t},A_{t-1},\mu,c_{w,t},c_{m,t},x_{t},P_{w,t})=v_{j,t}(\psi_{t},z_{w,t},z_{m,t},h_{w,t},A_{t-1},\mu,c_{w,t},c_{m,t},x_{t},P_{w,t})-V_{j,t}^{m\rightarrow s}(z_{j,t},\kappa_{j}A_{t-1})
$$
denote the marital surplus of household member $j$. The three cases
are

1. If $S_{j,t}(\mu_{t-1})\geq0$ for both $j=w,m$, they remain married
and keep the bargaining power unchanged. In turn, I  have $\mu_{t}=\mu_{t-1}$,
$(c_{w,t}^{\star},c_{m,t}^{\star},x_{t}^{\star},P_{w,t}^{\star})=(\tilde{c}_{w,t}(\mu_{t-1}),\tilde{c}_{m,t}(\mu_{t-1}),\tilde{x}_{t}(\mu_{t-1}),\tilde{P}_{w,t}(\mu_{t-1}))$,
and $D_{t}^{\star}=0$.
2. If $S_{j,t}(\mu_{t-1})<0$ for both $j=w,m$, they divorce. In turn
$D_{t}^{\star}=1$ and only $V_{j,t}^{m\rightarrow s}(z_{j,t}, A_{t-1})$ matters.
3. If one household member, say the woman, has a negative marital surplus
while the man has a positive marital surplus, they re-negotiate $\mu_{t}$.
They do so by finding the lowest value $\tilde{\mu}$ that solves $$ \tilde{\mu}:S_{w,t}(\tilde{\mu})=0 $$ 
making her just indifferent between remaining married and divorcing.
If the man also has a positive surplus for this value, $S_{m,t}(\tilde{\mu})\geq0$,
they remain married and increase the bargaining power of the woman.
In turn, $\mu_{t}=\tilde{\mu}$, $(c_{w,t}^{\star},c_{m,t}^{\star},x_{t}^{\star},P_{w,t}^{\star})=(\tilde{c}_{w,t}(\tilde{\mu}),\tilde{c}_{m,t}(\tilde{\mu}),\tilde{x}_{t}(\tilde{\mu}),\tilde{P}_{w,t}(\tilde{\mu}))$,
and $D_{t}^{\star}=0$. If, on the other hand, $S_{m,t}(\tilde{\mu})<0$,
there is no value of $\mu$ that can sustain the marriage, and the
couple sets $D_{t}^{\star}=1$.
